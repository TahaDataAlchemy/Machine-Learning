Scattering Plot using 2 variables

```python
#This is Bi variate
sns.scatterplot(x=df_tips["total_bill"], y=df_tips["tip"])
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_tips is your DataFrame with columns "total_bill," "tip," "sex," "smoke," and "size"
sns.scatterplot(x=df_tips["total_bill"], y=df_tips["tip"], hue=df_tips["sex"], style=df_tips["smoker"], size=df_tips["size"])
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```
