## Equation of line

In the equation \( y = mx + c \), \( m \) represents the slope of the line, and \( c \) represents the y-intercept.

- **Slope (\( m \))**: The slope (\( m \)) of a line describes the steepness or incline of the line. It indicates how much \( y \) changes for a unit change in \( x \). If \( m \) is positive, the line slopes upward to the right, indicating a positive relationship between \( x \) and \( y \). If \( m \) is negative, the line slopes downward to the right, indicating a negative relationship between \( x \) and \( y \).

- **Y-intercept (\( c \))**: The y-intercept (\( c \)) is the point where the line intersects the y-axis. It represents the value of \( y \) when \( x \) is 0. In other words, it's the value of \( y \) when \( x \) has no effect on the relationship between \( x \) and \( y \). It determines the starting point of the line on the y-axis.

Here's a simple example to illustrate the concepts:

Let's say we have a linear relationship between the number of hours studied (\( x \)) and the score obtained on a test (\( y \)). We want to predict the test score (\( y \)) based on the number of hours studied (\( x \)). 

- **Slope (\( m \))**: Suppose the slope (\( m \)) is 2. This means that for every additional hour studied (\( x \)), the test score (\( y \)) increases by 2 points. So, if a student studies for 3 hours (\( x = 3 \)), we would expect the predicted test score (\( y \)) to be \( 2 \times 3 = 6 \) points higher than if the student didn't study at all.

- **Y-intercept (\( c \))**: Suppose the y-intercept (\( c \)) is 5. This means that even if a student doesn't study at all (\( x = 0 \)), we would still expect the test score (\( y \)) to be 5 points (the y-intercept value). 

So, the equation \( y = 2x + 5 \) represents a linear model where the test score (\( y \)) is predicted based on the number of hours studied (\( x \)), with a slope of 2 (indicating that each additional hour studied results in a 2-point increase in the score) and a y-intercept of 5 (indicating that even if the student doesn't study, the expected score would be 5 points).

# GRADIENT DESCENT IMPLEMENTATION

We are implementing gradient descent to optimize the prediction process using the linear regression equation. This allows us to efficiently find the optimal parameters that minimize the prediction error, resulting in more accurate and reliable predictions.

The objective is to find the m and b (slope and y-intercept)  we have been given with x and y values 

learning rate in the codes resembles that how much optimal jump have to be taken by values to reach global minima


## HOW M AND C IS DERIVING

***Derivative with respect to m***: This tells us how the cost function changes as we change the slope of the regression line. If the derivative is positive, it means that increasing the slope will increase the cost function, so we need to decrease the slope. Conversely, if the derivative is negative, decreasing the slope will increase the cost function, so we need to increase the slope. By taking the derivative and updating m in the opposite direction of the derivative, we move closer to the optimal slope that minimizes the cost function.

***Derivative with respect to c***: Similarly, this tells us how the cost function changes as we change the intercept of the regression line. If the derivative is positive, increasing the intercept will increase the cost function, so we need to decrease the intercept. If the derivative is negative, decreasing the intercept will increase the cost function, so we need to increase the intercept. By taking the derivative and updating c in the opposite direction of the derivative, we move closer to the optimal intercept that minimizes the cost function.

```python
# EQ FORMAT WE ARE USING FOR LINE IS y = mx+b
def gradient_descent(x,y):
    m_curr = b_curr = 0 #choses randomly
    iterations = 10000 # randomly choses
    n = len(x)
    learning_rate = 0.08 #from where is cost start reducing
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr #line eq for predicting y
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)]) # cost eq
        md = -(2/n)*sum(x*(y-y_predicted)) # derivative eq for m(slope)
        bd = -(2/n)*sum(y-y_predicted) # derivative eq for b(intercept)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m {}, b{} ,cost {}, iteration {}".format(m_curr,b_curr,cost,i))
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
